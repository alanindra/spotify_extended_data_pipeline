## 1. Creating streaming table

```python
import pandas as pd
from pathlib import Path
import os

folder = Path('/Users/qaulanmaruf/Desktop/alans_spotify_2025/spotify_account_data')
tables = []

for file in folder.glob('StreamingHistory_music*'):
    tables.append(pd.read_json(file))

streaming_table = pd.concat(tables, ignore_index=True)
```
