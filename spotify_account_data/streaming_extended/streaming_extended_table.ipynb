## Most skipped artist

```python
query = """ 
    with
        track_skipped_status as (
            select 
                master_metadata_album_artist_name
                , skipped
                , count(*) count
            from 
                extended_stream_table
            where 
                ms_played <> 0
                and master_metadata_album_artist_name is not null
            group by 
                master_metadata_album_artist_name
                , skipped
        )

    select 
        master_metadata_album_artist_name
        , sum(case when skipped = 1 then count else 0 end) as count_skipped
        , sum(case when skipped = 0 then count else 0 end) as count_unskipped
        , (sum(case when skipped = 1 then count else 0 end))+(sum(case when skipped = 0 then count else 0 end)) count_total
        , (
            cast(sum(case when skipped = 1 then count else 0 end) as real)
            / (sum(count))
        ) as skip_rate
    from 
        track_skipped_status 
    group by 
        master_metadata_album_artist_name
    having 
        sum(count) > 250
    order by count_total desc
    ;
"""
```

```python
df = psql(query)
display(df.head(20))
```

## Top 10 listened artist in country of listening

```python
# top 5 countries

query = """ 
    select
        conn_country
        , count(*) count
    from 
        extended_stream_table
    where 
        ms_played <> 0
        and master_metadata_album_artist_name is not null
        and conn_country is not null
    group by 
        conn_country
    order by 
        count desc

    ;
"""
```

```python
df = psql(query)
print(df)
```

```python
query = """ 
    with
        ranked as (
            select
                conn_country
                , master_metadata_album_artist_name
                , count(*) count
                , (
                    cast(count(*) as real)
                    / sum(count(*)) over (partition by conn_country)
                ) as pct_in_country
                , row_number() over (
                    partition by conn_country order by count(*) desc
                ) as rn_helper
                , sum(count(*)) over (
                    partition by conn_country
                ) as country_total
            from 
                extended_stream_table
            where 
                ms_played <> 0
                and master_metadata_album_artist_name is not null
                and conn_country is not null
            group by
                master_metadata_album_artist_name
                , conn_country  
        )
    
    select
        conn_country
        , master_metadata_album_artist_name
        , count
        , pct_in_country
    from ranked 
    where 
        rn_helper <= 10
        and conn_country in (
            'ID','NL','AU','DE' -- countries i spent significant time in
        )
    order by 
        country_total desc,
        conn_country asc,
        count desc

    ;
"""
```

```python
df = psql(query)
display(df)
```

## Top 10 tracks in countries of listening

```python
query = """ 
    with
        ranked as (
            select
                conn_country
                , master_metadata_track_name
                , count(*) count
                , (
                    cast(count(*) as real)
                    / sum(count(*)) over (partition by conn_country)
                ) as pct_in_country
                , row_number() over (
                    partition by conn_country order by count(*) desc
                ) as rn_helper
                , sum(count(*)) over (
                    partition by conn_country
                ) as country_total
            from 
                extended_stream_table
            where 
                ms_played <> 0
                and master_metadata_track_name is not null
                and conn_country is not null
            group by
                master_metadata_track_name
                , conn_country  
        )
    
    select
        conn_country
        , master_metadata_track_name
        , count
        , pct_in_country
    from ranked 
    where 
        rn_helper <= 10
        and conn_country in (
            'ID','NL','AU','DE' -- countries i spent significant time in
        )
    order by 
        country_total desc,
        conn_country asc,
        count desc

    ;
"""
```

```python
df = psql(query)
display(df)
```
