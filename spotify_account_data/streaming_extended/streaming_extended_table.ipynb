## Creating extended audio streaming table

```python
import pandas as pd
from pathlib import Path
import os

FOLDER = Path('/Users/qaulanmaruf/Desktop/alans_spotify_2025/spotify_account_data/streaming_extended')
extended_audio_stream_table = []

for file in FOLDER.glob('Streaming_History_Audio*'):
    extended_audio_stream_table.append(pd.read_json(file))

extended_audio_stream_table = pd.concat(extended_audio_stream_table, ignore_index=True)
extended_audio_stream_table = extended_audio_stream_table.drop(columns=["ip_addr"])
```

## Extended video streaming table

```python
extended_video_stream_table = pd.read_json(FOLDER / "Streaming_History_Video_2022-2026.json")
extended_video_stream_table = extended_video_stream_table.drop(columns=["ip_addr"])
```

## Appending both tables as one
Assuming that video tables is part of streamed instances

```python
extended_stream_table = pd.concat([extended_audio_stream_table, extended_video_stream_table], ignore_index=True)
```

## Data validation
Exploring table feature using SQL

```python
from pandasql import sqldf

psql = lambda q: sqldf(q, globals())
```

```python
# min max ms_played

query = """ 
    select 
        min(ms_played) min_ms_played
        , max(ms_played) max_ms_played
    from(
        select distinct 
            ms_played 
        from 
            extended_stream_table
    )
    ;
"""

print(psql(query))
```

```python
# query = """ 
#     select distinct reason_start from extended_stream;
# """

query = """ 
    select distinct reason_end from extended_stream_table;
"""

print(psql(query))
```

```python
# FINAL QUERY

query = """ 
    select * from (
        select 
            ts
            , ms_played
            , master_metadata_track_name
            , master_metadata_album_album_name
            , master_metadata_album_artist_name
            , skipped
        from 
            extended_stream_table 
        where skipped = 1
    )
    where ms_played <> 0
    ;
"""

df = psql(query)

display(df.head(10))

# meaning need to filter out instances where songs are skipped
```
