# Creating streaming table

```python
import pandas as pd
from pathlib import Path
import os

FOLDER_STREAM = Path('/Users/qaulanmaruf/Desktop/alans_spotify_2025/spotify_account_data/streaming')
stream_table = []

for file in FOLDER_STREAM.glob('StreamingHistory_music*'):
    stream_table.append(pd.read_json(file))

stream_table = pd.concat(stream_table, ignore_index=True)
```

```python
from IPython.display import display

display(stream_table.head(10))
```

```python
from pandasql import sqldf

psql = lambda q: sqldf(q, globals())
```

```python
query = """ 
    select
        endTime end_ts
        , DATE(endTime, '+7 hours') AS end_date
        , STRFTIME('%H:%M', endTime, '+7 hours') AS end_time_hm
        , STRFTIME('%H', endTime, '+7 hours') AS end_time_h
    from 
        stream_table
    where msPlayed > 0
"""

print(psql(query))
```
